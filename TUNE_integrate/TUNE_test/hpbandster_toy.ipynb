```python


import logging
import hpbandster.core.nameserver as hpns
import hpbandster.core.result as hpres
from hpbandster.optimizers import BOHB as BOHB
from hpbandster.examples.commons import MyWorker

```

```python

NS = hpns.NameServer(run_id='example1', host='127.0.0.1', port=None)
NS.start()

```

```python


w = MyWorker(sleep_interval=0,nameserver='127.0.0.1',run_id='example1')
w.run(background=True)

```

```python


bohb = BOHB(configspace=w.get_configspace(),\
           run_id = 'example1', nameserver='127.0.0.1',\
           min_budget=9, max_budget=243)

res = bohb.run(n_iterations=4)
```

```python
bohb.shutdown(shutdown_workers=True)
NS.shutdown()
```

```python


id2config=res.get_id2config_mapping()
incumbent = res.get_incumbent_id()

print(id2config[incumbent]['config'],'best config')
print(len(id2config.keys()), 'total configs')
print(len(res.get_all_runs()), 'total runs')



print('%.1f'%(sum([r.budget for r in res.get_all_runs()])/243), 'total budget to full function evaluations')
```

```python


NS = hpns.NameServer(run_id='example2', host='127.0.0.1', port=None)
NS.start()

```

```python


workers=[]
for i in range(2):
    w = MyWorker(sleep_interval=0.5,nameserver='127.0.0.1',run_id='example2', id=i)
    w.run(background=True)
    workers.append(w)

```

```python
bohb = BOHB(configspace=w.get_configspace(),\
           run_id='example2',\
           min_budget=9,max_budget=243)
res = bohb.run(n_iterations=4,min_n_workers=2)
```

```python
bohb.shutdown(shutdown_workers=True)
NS.shutdown()
```

```python



id2config = res.get_id2config_mapping()
incumbent = res.get_incumbent_id()
all_runs = res.get_all_runs()

print(id2config[incumbent]['config'],len(id2config.keys()),len(all_runs))

print('Total budget corresponds to %.1f full function evaluations.'%(sum([r.budget for r in all_runs])/243))
print('The run took  %.1f seconds to complete.'%(all_runs[-1].time_stamps['finished'] - all_runs[0].time_stamps['started']))
```
